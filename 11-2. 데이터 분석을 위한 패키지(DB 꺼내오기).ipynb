## #drop 메서드 사용하여 데이터 삭제하기

```python
df_departments.drop(df_departments.index[[1,4]]).head()
# 1,4행 삭제
```

```python
df_departments.drop('DEPARTMENT_ID',axis=1)
# df_departments.drop(columns='DEPARTMENT_ID') 와 같은의미

# 2차원 배열에서
# axis=0 은 행(x축)을 기준으로 제거
# axis=1 은 열(y축)을 기준으로 제거
```

```python
df_departments.drop("DEPARTMENT_ID", axis=1, inplace=True)
df_departments
# Delete the Area column from the dataframe in place
# Note that the original 'data' object is changed 
# when inplace=True
```
