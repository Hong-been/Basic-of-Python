### NaN 처리방법
- 데이터에서 삭제 : dropna 함수 사용
- 다른값으로 치환 : fillna 함수 사용
- None vs NaN
    - None
        1. python 코드에서 누락 된 데이터에 자주 사용되는 Python singleton 개념이다.
        2. 숫자대신으로 사용될 수 없다.(at least shouldn't)
    
    - NaN
        1. 표준 IEEE 부동 소수점 표현을 사용하는 모든 시스템에서 인식되는 특수 부동 소수점 값을 의미한다.
        2. 'Not a Number' 을 의미하며 숫자대신 사용할 수 있다.
        3. isnan(2) returns False

```python
df_train.dropna()
# nan가 존재하는 행을 삭제
# df_train.dropna(axis=0, how='any') 와 같은 결과
# how='all'이면 전체가 NaN인 행만 삭제된다.(더 까다로운 삭제조건)
```

### #NaN 데이터를 고려하지 않아 오류가 나는 경우

### #NaN 데이터를 처리하여 오류 고치기
- math.isnan(nan) retuns True
- apple( function, axis)

```python
import math

def age_categorize(age):
    if math.isnan(age): #naN일 경우 -1로 바꿔줌
        return -1
    return int(age/10)*10 # 바닥값, 버림?
```

```python
df_train['Age'].apply(age_categorize) #naN처리를 하였기때문에 정상동작
```

```python
pclass_group=df_train.groupby('Pclass')
pclass_group.mean()[['Survived','Age']]
# 1등석의 승객이 가장 생존률이 높다.
# 3등석의 승객이 가장 생존률이 낮다.
```

```python
import numpy as np
df_train.groupby(['Pclass','Age']).aggregate([np.sum, np.mean, np.max])
# aggregate 집합
```

```python
df_train.groupby(['Pclass','Age']).aggregate([np.sum, np.mean,np.max])['Survived']
```

```python
df_train.groupby('Pclass').mean()
```

```python
df_train.groupby('Pclass').transform(np.mean)
# Pclass로 묶어서 나온 mean값을 다시 펼쳐서 형태롤 바꾼다.
```
