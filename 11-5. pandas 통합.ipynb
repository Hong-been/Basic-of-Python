## # append함수 사용하여 데이터 세로 통합하기

```python
df1=pd.DataFrame({'class1': [95,92,98,100],
                 'class2':[91,93,97,99]})
df1
```

```python
df2=pd.DataFrame({'class1': [87,89],
                 'class2':[85,90]})
df2
```

```python
df1.append(df2) # 새로운 데이터를 세로로 추가하기
```

```python
df1.append(df2,ignore_index=True) # 새로운 데이터 추가할 때 순차적으로 index가 증가하도록 옵션
```

```python
df3=pd.DataFrame({'class1':[96,83]})
df3
```

```python
df2.append(df3,ignore_index=True) # 새로운 데이터를 추가할때, 부족한 데이터는 NaN으로 입력된다.
```

## # merge 함수 사용하여 특정 열을 기준으로 통합하기
- how : left(왼쪽모두), right(오른쪽 모두), outer(양쪽모두),inner(양쪽 공통인 항목만, default)

```python
emp_df=pd.read_csv('employees.csv')
emp_df.head()
```

```python
dept_df=pd.read_csv('departments.csv')
dept_df.head()
```

```python
pd.merge(emp_df,dept_df, how='inner',on='DEPARTMENT_ID')
# 양쪽 모두 가지고 있는 DEPARTMENT_ID를 기준으로 통합한다.

#pd.merge(emp_df,dept_df, how='inner',on='DEPARTMENT_ID')['EMPLOYEE_ID'] 
# COLUMN을 두개이상하면 에러남..왜지ㅜ
```

```python
df_A_B=pd.DataFrame( {'판매월':['1월', '2월','3월','4월'],
                     '제품A':[100,150,200,140],
                     '제품B':[90,100,140,170]})
df_A_B
```

```python
df_C_D=pd.DataFrame( {'판매월':['1월', '2월','3월','4월'],
                     '제품C':[112,141,203,134],
                     '제품D':[90,110,140,170]})
df_C_D
```

```python
df_A_B.merge(df_C_D) # 공통된 column '판매월'을 기준으로 통합됨
```

## # concat 함수 사용하여 여러 데이터 통합하기

```python
result=pd.concat([df_A_B,df_C_D])
result
```
