```python
import pandas as pd
import numpy as np

df_cus=pd.read_csv('./data/customer.csv')
df_ord=pd.read_csv('./data/orders.csv')

df_cus.append(df_ord, ignore_index=True)
```

```python
# Q1. 가장 많이 팔린 물건은?
order_group=df_ord.groupby('item')
order_group['quantity'].sum()

# A1. 수건이 5개로 가장 많이 팔렸다.
```

```python
# Q2. 영희가 가장 많이 구매한 아이템은?
# A2. 치약을 4개로 가장 많이 구매하였다.

df_merge=df_cus.merge(df_ord,on='customer_id')
df_sum_quantity=df_merge[df_merge.name=='영희'].groupby('item')
df_sum_quantity['quantity'].sum()
```

```python
# A1. 티쳐's code
pd.merge(df_cus,df_ord,on='customer_id').groupby('item').sum().sort_values(by='quantity',ascending=False)
```

```python
# A2. 
pd.merge(df_cus,df_ord,on='customer_id').groupby(['name','item']).sum().sort_values(by='quantity',ascending=False)
# groupby() : 여러 column을 넣을 수 있다
# ascending= True 는 오름차순(default)

# sum().sort_va;es 함수는 표 형태로 예쁘게 나온다.
# ['quantity'].sum().sort_values 힘수는 거지같이 나온다....
# column이 하나면 by를 쓸 수 없고 거지같이 나옴!!
```
