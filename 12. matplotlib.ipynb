# 여러그래프 한 plot에 그리기

```python
y2=5*x+30
y3=4*x**2+10

plt.plot(x,y,x,y2,x,y3)
plt.show()
```

# 여러그래프 다른 plot에 그리기

```python
plt.plot(x,y)

plt.figure() # 새창에 따로그리기
plt.plot(x,y2)
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y1=x**2
y2=20*np.cos(x)**2

plt.figure(1) # 1번 창 생성
plt.plot(x,y1) # 1번 창에 그리기

plt.figure(2)  # 2번 창 생성
plt.plot(x,y2) # 2번창에 그리기

plt.figure(1) # 1번창 가져오기
plt.plot(x,y2) # 1번창에 그리기

plt.figure(2) # 2번창 가져오기
plt.clf() # 2번창 초기화
plt.plot(x,y1) # 2번창에 그리기

plt.show()
```

# 다양한 선 그래프 그리기

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1,'m') # 컬러 지정
plt.plot(x,y2,':') # 점선 지정
plt.plot(x,y3,'--') #막대선 지정
plt.show()
```

# 점 시각화 종류

```python
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d')
plt.show()
```

# 축, 그래프, grid 설명

```python
plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')

plt.xlabel('x axis') # 축 설명
plt.ylabel('y axis')
plt.title('Graph')
# 한글 입력 불가

plt.grid(True) # grid

plt.legend(['data1','data2','data3','data4'],loc='lower right') # 그래프 설명

plt.show()
```

# 도시별 인구밀도 그래프 그리기

```python
city=['서울','인천','대전','대구','울산','부산','광주']

lat=[37.56,37.45,36.35,35.87,35.53,35.18,35.16]
lon=[126.97,126.70,127.38,128.620,129.31,129.07,126.85]

pop_den=[16154,2751,2839,2790,1099,4454,2995]

color=['r','g','b','c','m','k','y']

size=np.array(pop_den)*0.3
# 인구 밀도에 따라 점 크기 설정


plt.scatter(lon,lat,s=size,c=color,alpha=0.5)

plt.xlabel('Longitude') # 축 설명
plt.ylabel('Latitude')
plt.title('Population by City')

for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
    
plt.grid()
```

# 막대그래프 bar


```python
member_id=['m1','m2','m3','m4']
before=[27,35,40,33]
after=[30,38,42,37]

index=np.arange(len(member_id))
plt.bar(index,before,color=color,tick_label=member_id,width=0.4)
# color인쓰면 에러, 순서 유관
plt.grid()
```

```python
plt.barh(index,before,color=color,tick_label=member_id,height=0.4)
# 세로막대는 barh
# weight 대신 height
plt.grid()
```

```python
barwidth=0.4
plt.bar(index,before,color='c',align='edge',width=barwidth,label='before')
plt.bar(index+barwidth,after,color='m',align='edge',width=barwidth,label='after')

plt.xticks(index+barwidth,member_id)
plt.legend()

plt.xlabel('회원 ID') # 축 설명
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전과 후의 근지구력 변화 비교')

plt.show()
```
